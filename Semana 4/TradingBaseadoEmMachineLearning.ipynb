# Trading Baseado em Machine Learning

Machine Learning é a combinação entre a ciência e a arte de programar computadores para que possam aprender com os dados.

A aprendizagem nesse contexto significa identificar padrões complexos dos dados.

O machine learning pode possuir característica de ser uma:
- Aprendizagem supervisionado: possui como propósito a classificação e a regressão;
- Aprendizagem não supervisionado: possui como propósito a agrupação(clustering) e redução de dimensão;

## Aprendizagem supervisionada
### Classificação
Buscam predizer se uma observação pertence a uma determinada categoria. 

No caso dos algoritmos de negociação, podemos prognosticar se um preço vai subir ou descer com base no histórico desse ativo.

Os principais algoritmos de classificação são:
- Regressão logística;
- Máquinas de suportes vetoriais;
- Árvores de classificação;
- Redes neurais.

### Regressão
Buscam predizer uma variável de saída baseado em um conjunto de dados.

No caso de algoritmos de negociação, poderíamos aplicar um modelo de regressão treinado com o histórico do ativo para predizer o preço futuro.

Alguns algoritmos de regressão são:
- Regressões lineares;
- Séries de tempos;
- Máquinas de suportes vetoriais;
- Árvores de decisões;
- Redes neurais.

## Aprendizagem não supervisionada

### Agrupamentos(clustering)

Buscam encontrar padrões ocultos e agrupamentos (cluster) dentro dos dados.

No caso de algoritmos de negociação, poderíamos aplicar o clustering para auxiliar no investimento a longo prazo para empresas de um determinado mercado. Em outras palavras, os algoritmos de clustering buscaram padrões ocultos para segmentar grupos de determinadas empresas para selecionar em qual investir.

O mais utilizado é o K-Médias.

### Redução de Dimensão

Tem como objetivo reduzir o número de variáveis, atributos ou características dos modelos.

No caso de algoritmos de negociação, com diversos indicadores e variáveis financeiras, poderíamos aplicar a redução para diminuir o número de variáveis sem que ocorra a perca de informação.

O algoritmo de redução mais utilizado é a análise de componentes principais.

## Trading Ativo

Busca negociação de curto prazo.

Os modelos mais utilizados de machine learning são modelos supervisionados de classificação.

Dado que a compra ou venda é uma decisão binária, pode-se associar essa decisão ao modelo de classificação, utilizando os indicadores técnicos e histórico de preços, treina-se os algoritmos de negociação para predizer se os preços subirão ou descerão.



# Principais Algoritmos de Negociação

## Máquinas de Suporte Vetorial (SVM)

São um modelo de machine learning que buscam uma classificação linear ou não linear.

O método baseia-se na maximização das margens de separação entre duas classes.

\begin{equation}
y = \{+1, -1\}
\end{equation}

Há uma função núcleo (Kernel) que maximiza a distância dos pontos de cada classe.

### Núcleo (Kernel)

A função do núcleo pode ser não linear, sendo uma função:
- Polinominal;
- Gaussiana;
- Exponencial;
- Laplaciano;
- Hiperbólico;
- Base radial.

A buscado núcleo ideal é o que define o sucesso do modelo.

## Árvores de Decisão

São algortimos versáteis. 

Utilizam de divisões binárias recursivas para classificar as observações.

### Random Forest (busca aleátoria)

É considerado o algortimo mais popular/poderoso de árvore de decisão.

Gerados 500 amostras diferentes da amostra inicial, treina-se cada amostra com uma árvore de decisão e, por último, junta-se todas as árvores em uma só.

A potência está no treino das subamostras.

## Redes Neurais (ANN)

São algoritmos baseados no funcionamento do cérebro humano.

Possuem três camadas:
- Entrada: na qual recebem os dados;
- Ocultas: em que os dados são processados;
- Saída: enviam a informação ao exterior;

No caso do trading ativo, podemos estruturar uma camada de entrada com indicadores técnicos, camadas ocultas para o processamento e uma última camada de saída com a informação se o preço sobe ou desce.

# Qual utilizar?

Para tomar a decisão, é necessário conhecer as métricas para avaliação de cada algoritmo.

## Conjuntos de Treino e Teste

Refere-se a divisão que é necessária ser feita para avaliar o algoritmo.

- **Conjunto de treino**: um subconjunto para treinar o modelo;
- **Conjunto de teste**: um subconjunto para testar o modelo treinado;

## Avaliando o Desempenho

Para a avaliação do desempenho do algoritmo, utiliza-se a **matriz de confusão**.

### Matriz de Confusão

É uma ferramento que permite a visualização do desempenho de um algoritmo de confusão.

|                  | Previu Venda -1             | Previu Compra      +1        |
|------------------|-----------------------------|------------------------------|
| **Venda Real (-1)**  | Verdadeiro Negativos (VN) | Falsos Positivos (FP)      |
| **Compra Real (+1)** | Falsos Negativos (FN)     | Verdadeiros Positivos (VP) |

A acurácia é uma medida para avaliar os modelos de classificação, é equivalente a proporção do número de predições corretas pelo total de predições.

\begin{equation*}
Acurácia = \frac{Número de Predições Corretas}{Total Predições}
\end{equation*}

Ou seja, 

\begin{equation*}
Acurácia = \frac{VN + VP}{VN + FP + FN + VP}
\end{equation*}

Além disso, é possível definir ainda a precisão para cada uma das classificações

\begin{equation*}
Precisão (Compra) = \frac{VP}{VP + FP}
\end{equation*}

Por último, temos ainda o _recall_, que é a proporção real dos que foram identificados corretamente
\begin{equation*}
Recall = \frac{VP}{VP + FN}
\end{equation*}

## Curva ROC (_Receiver Operating Characteristic Curve_)

É um gráfico que mostra o rendimento de um algoritmo de classificação.

A curva compara a taxa de verdadeiros positivos que busca-se maximizar e a taxa dee falsos positivos que busca-se minimizar:
\begin{equation*}
TPR = \frac{VP}{VP + FN}
\end{equation*}

## AUC (_Area Under Curve ROC_)

É uma medição agregada dos rendimentos em todos os cenários de classificação possíveis.

O AUC oscila de 0 a 1, um modelo em que 100% das predições são corretas possuem AUC = 1.

```python
import numpy as np
import pandas as pd
import yfinance as yfin
import matplotlib.pyplot as plt
import mplfinance as mpf
import ta
import pandas_datareader as pdr

# Importa a função de dividir os dados entre treino e teste
from sklearn.model_selection import train_test_split

# Importa função para normalizar os dados de entrada de treino e teste
from sklearn.preprocessing import StandardScaler

# Importa função do algoritmos de ML
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

# Importa funções para medir desempenho do algoritmo de ML
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_roc_curve
from sklearn.metrics import RocCurveDisplay
```

```python
def graficoEstrategia(df, shortMM, longMM):
    plt.figure(figsize = (20, 10))
    plt.plot(df['Close'], alpha = 0.6, label = 'Preços')

    plt.scatter(df.index, df['Compra'], color = 'green', marker = '^', label = 'Compra')
    plt.scatter(df.index, df['Venda'], color = 'crimson', marker = 'v', label = 'Venda')

    plt.title(f'Sinais de Compra e Venda, EMA {shortMM} & EMA {longMM}')
    plt.xlabel('Dia')
    plt.ylabel('Preço USD/bbl')
    plt.legend(loc = 'upper left')

    return plt


def graficoPatrimonio(ativo):
    plt.figure(figsize = (20, 10))
    plt.plot(ativo['Equity'])
    plt.title('Patrimônio - Estratégia de Médias Móveis Exponencias')
    plt.xlabel('Dia')
    plt.ylabel('Valor')
    plt.show()

def calculaParametros(df, risk_free):
    retornos = np.log(df['Equity']).diff().mean()
    retornos = retornos * 252
    #print(retornos)
    volatility = np.log(df['Equity']).diff().std()
    volatility = volatility * np.sqrt(252)
    #print(volatility)
    sharpe_ratio = (retornos - risk_free)/volatility
    
    return retornos, volatility, sharpe_ratio   
    
def estrategiaMediaMovel(ativo, inicio, fim, intervalo, shortMM, longMM):
    def algoritmo(dfTemp):
        n = len(dfTemp)       
        sinais = [0]

        for i in range(1, n):
            short = dfTemp['EMA_short'][i]
            long = dfTemp['EMA_long'][i]

            previous_short = dfTemp['EMA_short'][i-1]
            previous_long = dfTemp['EMA_long'][i-1]

            if previous_short < previous_long and short >= long:
                sinais.append(1)

            elif previous_short > previous_long and short <= long:
                sinais.append(-1)

            else:
                sinais.append(0)

        dfTemp['Sinais'] = sinais
        dfTemp['Compra'] = np.where(dfTemp['Sinais'] == 1, dfTemp['Close'], np.nan)
        dfTemp['Venda'] = np.where(dfTemp['Sinais'] == -1, dfTemp['Close'], np.nan)
        
        return dfTemp
        


    df = yfin.download(ativo, inicio, fim, interval = intervalo)
    df = df.dropna()
    df['EMA_short'] = df['Close'].ewm(span =shortMM, adjust = False).mean()
    df['EMA_long'] = df['Close'].ewm(span = longMM, adjust = False).mean()
    
    df = algoritmo(df)
    #resultado = grafico(df, shortMM, longMM)
    
    return df#, grafico
    
def backtestEstrategia(ativo, equity, TP, SL):
    '''Backtesting: Retorna o DataFrame com a variação de patrimônio, retorno, volatilidade e sharpe ratio da estratégia'''   
    
    N = len(ativo)
    equity = [equity]

    # Take Profit (lucro/stop gain) TP
    # Stop Loss SL

    pos = 0

    for i in range(1, N):

        equity.append(equity[i-1])

        if pos == 1:
            if ativo['Close'][i] >= preco*(1 + TP):
                pos = 0
                equity[i] *= (1 + TP)

            elif ativo['Close'][i] <= preco*(1 - SL):
                pos = 0
                equity[i] *= (1 - SL)


        elif pos == -1:
            if ativo['Close'][i] <= preco*(1 - TP):
                pos = 0
                equity[i] *= (1 + TP)

            elif ativo['Close'][i] >= preco*(1 + SL):
                pos = 0
                equity[i] *= (1 - SL)


        else:
            if ativo['Sinais'][i] != 0:
                pos = ativo['Sinais'][i]
                preco = ativo['Close'][i]
        #print(equity)   

    ativo['Equity'] = equity
    
    

    return ativo
    
```

## Funções desse módulo...

```python
def indicadoresTecnicos(df, listaMM):
    '''Gera as colunas com os dados dos indicadores técnicos baseados em um dataframe e uma lista de médias móveis'''
    # Definindo as médias móveis pelo preço do ativo para identificar se está acima ou abaixo
    for mm in listaMM:
        df[f'EMA_{mm}'] = ta.trend.ema_indicator(close = df['Close'], window = mm, fillna = True)/df['Close']
    
    # Índice de Força Relativa (RSI) 
    df['RSI'] = ta.momentum.rsi(close = df['Close'], fillna = True)
    
    # Alcance Médio Real (ATR)
    df['ATR'] =  ta.volatility.average_true_range(high = df['High'], low = df['Low'], close = df['Close'], fillna = True)
    
    # Williams Percent Range
    df['WR'] = ta.momentum.williams_r(high = df['High'], low = df['Low'], close = df['Close'], fillna = True)
    return df

def tratandoDF(df):
    '''Adiciona uma coluna target com a info de compra ou venda e ajusta o DF dropando algumas linhas nas quais os indicadores não foram calculados'''
    # Criando um vetor em target para definir com base nos preços de fechamento se o preço de encerramento foi maior ou menor que o anterior
    cl = np.array(df['Close'])
    target = np.where(cl[1:] > cl[0:-1], 1, -1)
    # Como não teremos o preço da última linha, podemos dropa-la
    df.drop(df.tail(1).index, inplace = True)
    df['Target'] = target
    
    # Eliminando linhas que não possuem alguns dados para indicadores técnicos
    df.drop(df.head(29).index, inplace = True)
    
    
    return df


def matrizEntradaSaida(df):
    '''Gera as matrizes de dados de entrada e saída com base no recorte dos indicadores e do campo alvo (venda ou compra)'''
    # Precisamos construir uma matriz que contenha os dados dos indicadores técnicos que geramos
    X = np.array(df.iloc[:, 6:-1]) 

    # Também construímos um vetor que contém as variáveis de saída
    Y = np.array(df['Target'])
    
    return X, Y

def organizaTesteTreino(matrizEntrada, matrizSaida, testSize):
    '''Organiza os dados de teste e treino, também normaliza os indicadores. Retorna um dicionário'''    
    X_train, X_test, Y_train, Y_test = train_test_split(matrizEntrada, matrizSaida, test_size = testSize, shuffle = False)
    ss = StandardScaler()
    ss = ss.fit(X_train)
    X_train = ss.transform(X_train)
    X_test = ss.transform(X_test)
    dados = {
        'entradaTreino': X_train,
        'entradaTeste': X_test,
        'saidaTreino': Y_train,
        'saidaTeste': Y_test
    }
    
    return dados



def classificador(algoritmo, dados):
    '''Dado um algoritmoe e os dados, a função treina o algoritmo com base nos dados e gera as métricas'''
    algoritmo = algoritmo.fit(dados['entradaTreino'], dados['saidaTreino'])
    predicao = algoritmo.predict(dados['entradaTeste'])
    
    accScore = accuracy_score(dados['saidaTeste'], predicao)
    confMatrix = confusion_matrix(dados['saidaTeste'], predicao)
    
    saida = {
        'predicao': predicao,
        'metricas':{
            'acuracia': accScore,
            'confusao': confMatrix
        }        
    }
    return saida

def ajustaSinaisDF(df, dados, clf):
    '''Ajusta a coluna de sinais de compra ou venda do DF baseado no retorno do classificador de ML'''
    N = len(dados['saidaTeste'])
    df = df.tail(N)
    df['Sinais'] = clf['predicao']
    return df


def TradingComMachineLearning(infoAtivo, infoTecnica, infoBacktest):
    df = yfin.download(infoAtivo['ativo'], infoAtivo['inicio'], infoAtivo['fim'], interval = infoAtivo['intervalo'])
    df = indicadoresTecnicos(df, infoTecnica['listaMedias'])
    df = tratandoDF(df)
    X, Y = matrizEntradaSaida(df)
    dados = organizaTesteTreino(X, Y, 0.1)
    resClassif = classificador(infoTecnica['algoritmoClassificador'], dados)
    df = ajustaSinaisDF(df, dados, resClassif)
    df = backtestEstrategia(df, infoBacktest['equity'], infoBacktest['takeProfit'], infoBacktest['stopLoss'])
    saida = {
        'Estrategia': infoTecnica['estrategia'],
        'Informacoes': df,
        'desempenhoClassificador': resClassif,
        'dados': dados,
        'algoritmo': infoTecnica['algoritmoClassificador']
    }
    return saida
    
```

```python
infoAtivo = {
    'ativo': 'ETH-USD',
    'inicio': '2019-09-01',
    'fim': '2021-09-02',
    'intervalo': '1d'
}
infoBacktest ={
    'equity': 100,
    'takeProfit': 0.03,
    'stopLoss': 0.01
}
infoTecnicaRF = {
    'estrategia': 'Random Forest',
    'listaMedias': [5, 20, 50, 100],
    'pesoTeste': 0.1,
    'algoritmoClassificador': RandomForestClassifier()    
}
infoTecnicaSVM = {
    'estrategia': 'SVM',
    'listaMedias': [5, 20, 50, 100],
    'pesoTeste': 0.1,
    'algoritmoClassificador': SVC()    
}
infoTecnicaMLPC = {
    'estrategia': 'Rede Neural',
    'listaMedias': [5, 20, 50, 100],
    'pesoTeste': 0.1,
    'algoritmoClassificador': MLPClassifier(hidden_layer_sizes = (64, 32), max_iter = 1000)
}

ethereumRF = TradingComMachineLearning(infoAtivo, infoTecnicaRF, infoBacktest)
ethereumSVM = TradingComMachineLearning(infoAtivo, infoTecnicaSVM, infoBacktest)
ethereumMLPC = TradingComMachineLearning(infoAtivo, infoTecnicaMLPC, infoBacktest)
#ddApelaoRF = TradingComMachineLearning(infoAtivo, infoTecnicaRF, infoBacktest)
#ddApelaoMLPC = TradingComMachineLearning(infoAtivo, infoTecnicaMLPC, infoBacktest)
# btc_svm = TradingComMachineLearning(infoAtivo, infoTecnicaSVM, infoBacktest)
# btc_mlpc = TradingComMachineLearning(infoAtivo, infoTecnicaMLPC, infoBacktest)

#btc_estrategias = [ddApelaoSVM, ddApelaoRF, ddApelaoMPLC]
```
